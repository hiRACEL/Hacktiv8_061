## Case 3 : Top 5 Negara dengan Imigrasi terbanyak ke canada selamaa 1980 - 2013

```python
#buat df baru yang di sortit menurut data kolom total
df_can_sorted=df_can.sort_values(by='Total', ascending=False, axis=0)
df_top5 = df_can_sorted.head(5)
df_top5_years = df_top5.loc[:,years]
df_top5_years.T.plot()
```

```python
# karena plotnya kecil maka ditambahkan figsize / figure size atau "ukuran gambar

df_top5_years.T.plot(figsize = (15,10))
```

```python
df_top5_years.T.plot(marker='o',figsize = (15,10))

#pakai si matplotlib.pyplot untuk menambahakkan labelplt.title("Imigrasi Penduduk Haiti ke Canada")
plt.title ("Top 5 Negara yang melakukan imigrasi ke Canada")
plt.ylabel("Jumlah Imigran")
plt.xlabel ("Tahun")

plt.show()
```

## Area plot

```python
df_top5_years.T.plot(kind="area",figsize = (15,10))
```

```python
# kalo datanya tidak mau di tumpul , kita tambahkan stacked = Flase
df_top5_years.T.plot(kind="area",stacked=False, figsize = (15,10))

plt.title ('Top 5 Imigration to Canada 1990 - 2013')
plt.xlabel ('Years')
plt.ylabel ('Number of Imigrant')
```

```python
#perhatikan kalau warnanya agak2 transparan
#sebeeapa transparan warnanya bisa kita atur dengan opsi alpha = 0.25 (0%)

df_top5_years.T.plot(kind="area",stacked=False, figsize = (20,6), alpha=0.25)
```

```python
## Case 4 : Mmebuat area plot untuk 5 negara yang imigrasi paling sedikit ke canada selama 1980 -2013
# menggunakan scripting layer dan type  areanya stacked dan transparansinya 45%

df_can_sorted_asc=df_can.sort_values(by='Total', ascending=False, axis=0)
df_top5_lowest = df_can_sorted_asc.tail(5)
df_top5_years_lowest = df_top5_lowest.loc[:,years]
df_top5_years_lowest.T.plot(kind="area", stacked='True', alpha= 0.45, figsize=(20,6))

plt.title("5 Negara yang Imigrasi ke Canada paling sedikit 1990 - 2013")
plt.xlabel("Tahun")
plt.ylabel("Jumlah Imigran")
```

## Case 6 : Histogram - Freqkuensi distribusi data

Bsgaimana frekuensi distribusi jumlah imigrasi ke Canada dari setiap negara di tahun 2013



```python
df_can[2013]
```

```python
# pakai bantuan numpy.histogram untuk mencari 2 set data yabg kita perlukan

count, bin_edges = np.histogram(df_can[2013])

print(count)
print(bin_edges)
```

```python
#sebenrnya utk plot histogram kita bisa saja langsung plot dr data kita
#kasin kind='hist'

df_can[2013].plot(kind='hist')
```

```python
# krn garis diatas tidak pas maka menggunakan numpy.histogram tadi

df_can[2013].plot(kind='hist', xticks=bin_edges, figsize=(10,5))
```

```python
imigrant_2013 = df_can[2013].plot(kind='hist', xticks=bin_edges, figsize=(10,5))

imigrant_2013.set_title("Distirbusi jumlah imigran tahun 2013")
imigrant_2013.set_xlabel("Tahun")
imigrant_2013.set_ylabel("Jumlah Imigran")
```

```python
## Case 6 : Histogram - Freqkuensi distribusi data

#Bsgaimana frekuensi distribusi jumlah imigrasi ke Canada dari Denmar, Norweign., Swedden negara di tahun 1990-2013

df_dns= df_can.loc[['Denmark','Norway', 'Sweden'],years]
```

```python
df_dns = df_dns.T
df_dns.plot(kind='hist')
```

```python
df_dns2 = df_dns.T
dns = df_dns2.plot(kind='hist')

dns.set_xlabel("Tahun")
dns.set_ylabel("Jumlah Imigran")
```

```python
count, bin_edges = np.histogram(df_dns2,15)

print(count,len(count))
print(bin_edges)
```

```python
df_dns2 = df_dns.T
dns = df_dns2.plot(figsize=(10,5),kind='hist', bins=15,xticks=bin_edges, alpha = 0.45)

dns.set_title("Histogram Imigrasi dari Denmar, Norway, Swedden ke Canada")
dns.set_xlabel("Jumlah Imigran")
dns.set_ylabel("Jumlah Tahun")
```
